## This script randomly generates locations to be used in a quadratic assignment problem (QAP)
It will generate random coordinates on the unit circle scaled up by 10 times, and then write those as x, y pairings to a 2, n dim csv file 

```python
import random
import matplotlib.pyplot as plot
```

```python
# Defining crucial variables
n = 25 # number of locations to generate
circleSize = 25 # size of circle radius
```

```python
# Initializing arrays and creating random coords
x = []
y = []
for _ in range(n):
    x.append(random.uniform(-1,1)*circleSize)
    y.append(random.uniform(-1,1)*circleSize)
```

```python
# Plot positions
plot.scatter(x, y)
plot.show()
```

```python
# Write to csv
with open('locations.csv','w') as file:
    for index in range(len(x)):
        string = str(x[index]) + ',' + str(y[index])
        file.write(string)
        file.write('\n')
```
